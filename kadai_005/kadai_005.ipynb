```python
import matplotlib.pyplot as plt
import numpy as np

# プロットの作成
fig, axes = plt.subplots(2, 3, tight_layout=True)

# データの準備
y0 = [1, 2, -5, 2]
x0 = np.linspace(0, 10, 100)
y1 = np.sin(x0)
x1 = np.random.rand(100)
x2 = np.random.normal(15, 5, 2000)
y3 = np.random.rand(50)
x3 = np.random.rand(50)

# 折れ線グラフ表示
axes[0,0].plot(y0)
# sin関数表示
axes[0,1].plot(x0,y1)
# ヒストグラム
axes[0,2].hist(x2)
# 散布図
axes[1,0].scatter(x3,y3)

plt.show()
```
